```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/sample_data/sonar_dataset.csv', header=None)
```

```python
data.head()
```

```python
data.columns
```

```python
x = data.drop(60 , axis=1)
y = data[60]
```

```python
x
```

```python
y
```

```python
y = pd.get_dummies(y , drop_first=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
import tensorflow as tf
from tensorflow import keras
```

**BUILDING A MODEL WITHOUT DROP OUT LAYER AND CHECKING ACCURACY.**

```python
model = keras.Sequential([
      keras.layers.Dense(60, input_dim=(60) , activation='relu'),
      keras.layers.Dense(30,  activation='relu'),
      keras.layers.Dense(10,  activation='relu'),
      keras.layers.Dense(1,  activation='sigmoid'),
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train , y_train , epochs = 100 , batch_size=8)
```

```python
model.evaluate(x_test , y_test)
```

**BUILDING A MODEL WITH DROP OUT LAYERS AND CHECKING ACCURACY**

```python
model_d = keras.Sequential([
          keras.layers.Dense(60 , input_dim = (60) , activation='relu'),
          keras.layers.Dropout(0.5),
          keras.layers.Dense(30 , activation='relu'),
          keras.layers.Dropout(0.5),
          keras.layers.Dense(15 , activation='relu'),
          keras.layers.Dropout(0.5),
          keras.layers.Dense(1 , activation='sigmoid')
])

model_d.compile(
    optimizer ='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model_d.fit(x_train , y_train , epochs = 100 , batch_size=8)
```

```python
model_d.evaluate(x_test , y_test)
```

**TAKE PRECISION REPORT OF BOTH MODEL**

```python
y_predicted = model.predict(x_test).reshape(-1)
y_predicted = np.round(y_predicted)
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test , y_predicted))
```

```python
y_predicted_d = model_d.predict(x_test).reshape(-1)
y_predicted_d = np.round(y_predicted_d)

print(classification_report(y_test , y_predicted_d))
```
